```python
#Chapter 8 exercise 4
def any_lowercase1(s):
    for c in s:
        if c.islower():
            return True
        else:
            return False
any_lowercase1("TESt")
#only looks at the first letter and returns true if it is lowercase

def any_lowercase2(s):
    for c in s:
        if 'c'.islower():
            return 'True'
        else:
            return 'False'
any_lowercase2("TESt")
#the quotes around "c" in the islower function make it so the function only looks at the c

def any_lowercase3(s):
    for c in s:
        flag = c.islower()
    return flag
any_lowercase3("tEST")
#this one only returns if the last letter of the word is lowercase

def any_lowercase4(s):
    flag = False
    for c in s:
        flag = flag or c.islower()
    return flag
any_lowercase4("TEsT")
#this returns if any letter in the word is lowercase

def any_lowercase5(s):
    for c in s:
        if not c.islower():
            return False
    return True
any_lowercase5("test")
#this returns if any letter in the word is not lowercase
```
